```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
result, continuation_token = reviews(
    'blibli.mobile.commerce',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1200
)
```

```python
df = pd.DataFrame(result)

df.head()
```

```python
df = df[['content', 'score']]

df.head()
```
