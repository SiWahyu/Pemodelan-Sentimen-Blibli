## 2. Import Library

```python
import re
import joblib
import nltk


from nltk.corpus import stopwords

nltk.download('stopwords')
```

## 3. Load Model

```python
import joblib

model = joblib.load('/content/drive/MyDrive/svm_model.pkl')
vectorizer = joblib.load('/content/drive/MyDrive/tfidf_vectorizer.pkl')
```

## 4. Preprocessing

```python
stop_words = set(stopwords.words('indonesian'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    words = text.split()
    words = [word for word in words if word not in stop_words]

    return ' '.join(words)
```

## 5. Input Review

```python
review_baru = [
    "Aplikasi sangat bagus dan membantu proses belanja",
    "Customer Service nya kurang respon rada kecewa",
    "Lumayan oke"
]
```

## 6. Preprocessing Review

```python
review_clean = [preprocess(review) for review in review_baru]
```

## 7. TF-IDF

```python
review_tfidf = vectorizer.transform(review_clean)
```

## 8. Prediksi

```python
prediksi = model.predict(review_tfidf)
```

## 9. Hasil

```python
for review, hasil in zip(review_baru, prediksi):
    print(f"Review    : {review}")
    print(f"Sentimen  : {hasil}")
    print("-" * 50)
```
